<h2>
3. Revisando FACTURAS
</h2>

```python
FACTURAS_RAW = pd.read_excel('Comercializadora.xlsx', sheet_name='Facturas')
FACTURAS_RAW.head()
```

las columnas tienen acentos y espacios

```python
FACTURAS_RAW.columns.values
```

```python
ChngVector=[]
for variable in FACTURAS_RAW.columns.values:
    #Pascal Case
    chng = [i.title() for i in variable.split(' ')]
    ans=''
    for i in chng:
        ans = ans+ ''+ i
    variable = ans
    ChVal=mo.WebCleanText(variable).replace(' ','')
    ChngVector.append(ChVal)
FACTURAS_RAW.columns = ChngVector
```

```python
FACTURAS_RAW.columns
```

verificamos tipos de datos

```python
FACTURAS_RAW.describe(include = 'all').T
```

```python
FACTURAS_RAW.dtypes
```

no hay errores en tipos de variables

### Normalizacion de caracteres

es necesario excluir variables de fechas

```python
FACTURAS_RAW.columns
```

```python
columnas = ['Serie', 'Folio', 'RazonSocial', 'NombreDeLaMoneda', 'Total', 'TipoDeCambio', 'Cancelado', 'Referencia', 'CodigoDelAgente', 'Pendiente', 'NombreDelAgente', 'Impuesto1']
```

```python
for variable in columnas:
    ChngVector=[]
    for each in FACTURAS_RAW[variable]:
        ChVal=mo.WebCleanText(each)
        try: ChVal.replace("\\", " ")
        except: None
        ChngVector.append(ChVal)
    FACTURAS_RAW[variable] = ChngVector
FACTURAS_RAW.head(10)
```

```python
FACTURAS_RAW.tail(10)
```

Impuesto1 es un cargo numerico, no puede haber nulos, solo ceros

```python
NumVars=['Total','TipoDeCambio','Cancelado','Pendiente','Impuesto1']
```

```python
for i in NumVars:
    print(i,' ',FACTURAS_RAW[i].isnull().sum() )
```

```python
for i in NumVars:
    try: FACTURAS_RAW[i].replace('NaN', 0, inplace=True)
    except: print(i,' oops no es texto')
    try: FACTURAS_RAW[i].replace(np.nan, 0, inplace=True)
    except: print(i,' oops no es nulo')
```

```python
try: FACTURAS_RAW['Referencia'].replace('NaN', 0, inplace=True)
except: print(' oops no es texto')
```
